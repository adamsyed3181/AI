```python
from typing_extensions import overload
from collections import deque


def DFS(graph, root, end_goal): #creates DFS algorithm with a end goal target

    visited, queue = set(), deque([root])
    visited.add(root)

    while queue:
        vertex = queue.pop()
        print(vertex, end=" ")

        if vertex == end_goal: #if vertex is equal to end goal, loop ends and returns success
            return "SUCCESS"

        for neighbor in reversed(graph[vertex]):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)

    return "FAILURE" #if loop is empty and no end goal is reached, return failure

maze1 = {
  "S": ["1", "3"],
  "1": ["S", "2"],
  "2": ["1", "4"],
  "4": ["2", "5"],
  "5": ["4"],
  "3": ["S", "6"],
  "6": ["3", "7"],
  "7": ["6", "8"],
  "8": ["7", "9", "10"],
  "9": ["8", "11"],
  "11": ["9", "12"],
  "12": ["11"],
  "10": ["8", "13"],
  "13": ["10", "14"],
  "14": ["13", "15"],
  "15": ["14", "16"],
  "16": ["15", "17"],
  "17": ["16", "18"],
  "18": ["17", "19", "20"],
  "20": ["18", "21"],
  "21": ["20"],
  "19": ["18", "22", "23"],
  "22": ["19", "24"],
  "24": ["22", "25"],
  "25": ["24"],
  "23": ["19", "26"],
  "26": ["23", "27"],
  "27": ["26", "28"],
  "28": ["27", "29"],
  "29": ["28", "30"],
  "30": ["29", "31"],
  "31": ["30", "32", "33"],
  "33": ["31", "34"],
  "34": ["33", "35"],
  "35": ["34"],
  "32": ["31", "36"],
  "36": ["32", "37"],
  "37": ["36", "38"],
  "38": ["37", "39"],
  "39": ["38", "40"],
  "40": ["39", "41"],
  "41": ["40", "42"],
  "42": ["41", "43", "44"],
  "44": ["42", "45"],
  "45": ["44"],
  "43": ["42", "46"],
  "46": ["43", "47", "48"],
  "47": ["46"],
  "48": ["46", "G"],
  "G": ["48"]
        }
DFS(maze1, "S", "G")
```
